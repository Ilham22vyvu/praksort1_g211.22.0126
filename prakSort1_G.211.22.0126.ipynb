**Bubble Sort**

```python
def bubbleSort(arr):
    n = len(arr)
    swapped = False

    for i in range(n-1):
        for j in range(0, n-i-1):
            if arr[j] > arr[j + 1]:
                swapped = True
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
```

```python
# Driver code to test above
arr = [50, 54, 31, 29, 80, 43,76]
bubbleSort(arr)
print("Sorted array is:")
for i in range(len(arr)):
    print("% d" % arr[i], end=" ")
```

**Selection Sort**

```python
import sys
A = [69, 70, 21, 54, 34]
for i in range(len(A)):
    min_idx = i
    for j in range(i+1, len(A)):
        if A[min_idx] > A[j]:
            min_idx = j

        A[i], A[min_idx] = A[min_idx], A[i]
print ("Sorted array")
for i in range(len(A)):
 print("%d" %A[i],end=" ")
```

**Insertion Sort**

```python
def insertionSort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
    while j >= 0 and key < arr[j] :
        arr[j + 1] = arr[j]
        j -= 1
    arr[j + 1] = key
```

```python
# Driver code to test above
arr = [34, 21, 54, 65, 20]
insertionSort(arr)
for i in range(len(arr)):
    print ("% d" % arr[i])
```
